3. What is the regional utilization rate (RUR) for orders shipped? (RUR is defined as orders shipped in same zone)


```python
ecom_dataset.head()
source_zones.head()
```

```python
ecom_dataset_1=pd.merge(ecom_dataset,source_zones, how='left',left_on='ShippedFrom', right_on='ShippedFrom',suffixes=('_shipped','_shipped'))
```

```python
ecom_dataset_2=pd.merge(ecom_dataset_1,dest_zones, how='left',left_on='Country', right_on='DestinationCountry',suffixes=('_shipped','_dest'))
```

```python
ecom_dataset_2.drop(columns='DestinationCountry')
```

What is the regional utilization rate (RUR) for orders shipped? (RUR is defined as orders shipped in same zone)

```python
ecom_dataset_2[ecom_dataset_2.Zones_shipped==ecom_dataset_2.Zones_dest]
```

```python
len(ecom_dataset_2[ecom_dataset_2.Zones_dest==ecom_dataset_2.Zones_shipped])*100/len(ecom_dataset_2)
```

Can the RUR be improved? To what extent? Which country has best RUR?

```python
RUR_by_country=pd.DataFrame(ecom_dataset_2[ecom_dataset_2.Zones_dest==ecom_dataset_2.Zones_shipped].groupby('Country')['StockCode'].count()/ecom_dataset_2.groupby('Country')['StockCode'].count())
```

```python
RUR_by_country.unstack().sort_values(ascending=False)
```
